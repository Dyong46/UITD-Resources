### 1.2 Viết chương trình nhập vào 2 số nguyên a, b. Tính tổng, hiệu, tích, thương, chia lấy dư, chia lấy nguyên của 2 số trên và in kết quả ra màn hình. Kết quả phép chia làm tròn 2, 3 chữ số sau dấu chấm (ví dụ kết quả 3.333333 thì làm tròn 3.333).

```python
def number (a, b):
    sum = a + b
    sub = a - b
    mult = a * b
    div = round(a / b, 2)
    mod = a % b
    div_round = a // b
    return sum, sub, mult, div, mod, div_round
```

```python
number(7,3)
```

### 1.12 Viết chương trình xuất ra số (nguyên, thực) ngẫu nhiên theo yêu cầu sau:
• O đến 100
• 50 đến 99
• -39 đến 79
• -79 đến -39

```python
def random_number (val):
    if val == 1:
        return random.randint(0, 100)
    elif val == 2:
        return random.randint(50, 99)
    elif val == 3:
        return random.randint(-39, 79)
    elif val == 4:
        return random.randint(-79, -39)
    return None
```

```python
print('[0 - 100]: ', random_number(1))
print('[50 - 99]: ', random_number(2))
print('[-39 - 79]: ', random_number(3))
print('[-79 - -39]: ', random_number(4))
```

### 1.14 Viết chương trình nhập giá trị cho hai số thực a, b. Tính giá trị biểu thức sau:

```python
def cal_formula_1 (a, b):
    sqrt_a = math.sqrt(a)
    sqrt_b = math.sqrt(b)
    fourth_root_a = a ** 0.25
    fourth_root_b = b ** 0.25
    fourth_root_ab = (a * b) ** 0.25

    term_1 = (sqrt_a - sqrt_b) / (fourth_root_a - fourth_root_b)
    term_2 = (sqrt_a + fourth_root_ab) / (fourth_root_a + fourth_root_b)

    return term_1 - term_2
```

```python
cal_formula_1(16, 4)
```

```python
def cal_formula_2 (a, b):
    cbrt_a = a ** (1/3)
    cbrt_b = b ** (1/3)
    cbrt_ab = (a * b) ** (1/3)

    numerator = (a + b) / (cbrt_a + cbrt_b) - cbrt_ab
    denominator = (cbrt_a - cbrt_b) ** 2


    return numerator / denominator
```

```python
cal_formula_2(2, 3)
```

• Đại học Quốc gia

2. Tách thành các sub-string:

• Thủ Đức
• HCM

```python
address = 'Đại học Quốc gia, Khu phố 6, P. Linh Trung, Q. Thủ Đức, Tp. HCM'
```

```python
sub_string_1 = address.split(',')
sub_string_1 = [item.strip() for item in sub_string_1]
sub_string_1
```

```python
parts = [item.strip() for item in address.split(',')]

regex = r'(?:P\.|Q\.|Tp\.)\s*'

sub_string_2 = [re.sub(regex, '', part) for part in parts]

sub_string_2
```
