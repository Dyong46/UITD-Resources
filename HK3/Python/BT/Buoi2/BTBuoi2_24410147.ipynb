## 1. Rock - Paper - Scissor

```python
import random as r

choices = ['keo', 'bua', 'bao']

person = input("Person take (keo/bua/bao): ").strip().lower()
computer = r.choice(choices)

def keo_bua_bao(person, computer):
    if person not in choices:
        return "Invalid selection!"

    if person == computer:
        return 'Hòa'

    winning_cases = {
        'keo': 'bao',
        'bua': 'keo',
        'bao': 'bua'
    }

    if winning_cases[person] == computer:
        return 'Person win'
    else:
        return 'Computer win'

print('Máy ra:', computer)
print('Kết quả:', keo_bua_bao(person, computer))
```

## 2. Check the time

```python
t = input("The time (hh:mm:ss):")

l = t.split(":")

if len(l) != 3:
    print("Invalid time format!")
else:
    try:
        gio = int(l[0])
        phut = int(l[1])
        sec = int(l[2])

        if 0 <= gio <= 23 and 0 <= phut <= 59 and 0 <= sec <= 59:
            print("Valid time format!")
        else:
            print("Invalid time format!")
    except ValueError:
        print("Invalid time format!")
```
